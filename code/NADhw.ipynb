```python
import pandas as pd
import lzma

compressed_file = "../data/state_county/va_fairfax.csv.xz"
with lzma.open(compressed_file, mode='rt') as file:
    df = pd.read_csv(file)

output_file = "../data/state_county/va_fairfax.csv"
df.to_csv(output_file, index=False)


print('Conversion complete.')

```
